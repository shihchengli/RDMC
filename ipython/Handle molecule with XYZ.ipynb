This notebook demonstrates how RDMC handles XYZ. Molecules can be perceived by using [openbabel](http://openbabel.org/wiki/Main_Page) / [method from Jensen et al.](https://github.com/jensengroup/xyz2mol). The created molecule can be visualized by `mol_viewer` and exported to xyz format.

```python
from rdmc.mol import RDKitMol
from rdmc.view import mol_viewer
from rdkit.Chem.Draw import IPythonConsole
```

## 2. Generate an RDKitMol from xyz
**Arguments:**
- `backend`: choose the `openbabel` or `jensen` backend
- `header`: indicate if the str includes lines of atom number and title

```python
RDKitMol.FromXYZ(xyz, backend='openbabel', header=True)
```

Next is an example of parsing XYZ without the first two lines (e.g., extracted from gaussian output), where `header` is set to `False`. Please set the `header` arguments correctly according to the input xyz.

```python
RDKitMol.FromXYZ(xyz_without_header, backend='openbabel', header=False)
```

Use `jensen` backend. For most cases, Jensen's method returns the same molecule as using `openbabel` backend

```python
RDKitMol.FromXYZ(xyz, backend='jensen', header=True)
```

The optional arguments for the Jensen et al. method are provided.

```python
RDKitMol.FromXYZ(xyz,
                 backend='jensen',
                 header=True,
                 allow_charged_fragments=False,  # radical => False
                 use_graph=True,  # Use networkx to accelerate (enabled by default)
                 use_huckel=True,  # Use extended Huckel bond orders to locate bonds
                 embed_chiral=True,  # Embed chiral information
                 force_rdmc=False,  # use RDMC's implementation of xyz2mol
                 )
```

## 3. Check the coordinates of the molecule and visualize it

```python
mol = RDKitMol.FromXYZ(xyz)
print(f'xyz coordinate: \n{mol.GetPositions(id=0)}')
```

## 4. Export XYZ

```python
print(mol.ToXYZ(header=False))
```
